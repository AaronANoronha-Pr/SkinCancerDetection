```python
# Reading the data from HAM_metadata.csv
meta = pd.read_csv("//home//aaron//Downloads//archive//HAM10000_metadata.csv")
print(meta.head())
print(meta.shape)
```

```python
# Plotting the disease class frequencies
sns.countplot(x = 'dx', data = meta)
plt.xlabel('Disease', size=12)
plt.ylabel('Frequency', size=12)
plt.title('Frequency Distribution of Classes', size=16)
plt.show()
```

```python
# Age Frequencies
bar, ax = plt.subplots(figsize=(10, 10))
sns.histplot(meta['age'])
plt.title('Histogram of Age of Patients', size=16)
plt.show()
```

```python
# Location of diseases
location = meta['localization'].value_counts()
plt.figure(figsize=(20, 5))
sns.countplot(x='localization', data=meta)
plt.title('Disease localisation distribution')
plt.show()
```

```python
# Gender of patients
bar, ax = plt.subplots(figsize=(10, 10))
plt.pie(meta['sex'].value_counts(),
        labels = meta['sex'].value_counts().index, autopct="%.1f%%")
plt.title('Gender of Patient', size=16)
plt.show()
```
